```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
alexa_countvectorizer = vectorizer.fit_transform(alexa_data['verified_reviews'])
```

```python
alexa_countvectorizer.shape
```

```python
print(vectorizer.get_feature_names())
```

```python
print(alexa_countvectorizer.toarray())
```

```python
alexa_data.drop('verified_reviews', axis =1, inplace = True)
```

```python
#alexa_data = pd.concat([alexa_data,alexa_countvectorizer], axis = 1)
```

```python
encoded_reviews = pd.DataFrame(alexa_countvectorizer.toarray())
```

```python
alexa_data = pd.concat([alexa_data,encoded_reviews], axis = 1)
```

```python
alexa_data.head()
```

```python
X = alexa_data.drop('feedback', axis =1)
```

```python
X.shape
```

```python
y = alexa_data['feedback']
```

```python
y.shape
```

```python
#y.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=5)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
random_forest_classifier = RandomForestClassifier(n_estimators=100, criterion= 'entropy')
```

```python
random_forest_classifier.fit(X_train,y_train)
```

```python
y_pred = random_forest_classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm, annot= True)
print(classification_report(y_test,y_pred))
```

```python
alexa_df = pd.read_csv("amazon_alexa.tsv", sep ='\t')
```

```python
alexa_df = pd.concat([alexa_df, pd.DataFrame(alexa_countvectorizer.toarray())], axis = 1)
```

```python
alexa_df
```

```python
alexa_df['review_length'] = alexa_df['verified_reviews'].apply(len)
```

```python
alexa_df
```

```python
X = alexa_df.drop(['rating','date','variation','verified_reviews','feedback'], axis = 1)
```

```python
X
```

```python
y = alexa_df['feedback']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=5)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
random_forest_classifier_improved = RandomForestClassifier(n_estimators=100, criterion= 'entropy')
```

```python
random_forest_classifier_improved.fit(X_train,y_train)
```

```python
y_pred_improved = random_forest_classifier_improved.predict(X_test)
```

```python
cm_improved = confusion_matrix(y_test,y_pred_improved)
sns.heatmap(cm_improved, annot = True)
print(classification_report(y_test,y_pred_improved))
```
